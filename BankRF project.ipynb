```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
path = "bank_1.csv"
data=pd.read_csv(path)
data.head()
```

```python
data.tail()
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
data.columns
```

```python
selected_col=['default','housing','loan','deposit']
selected_col
```

```python
for col in selected_col:
    data[col]=data[col].map({"yes":1,"no":0})
```

```python
data.head()
```

```python
data=pd.get_dummies(data,columns=["job","marital","education","contact","month","poutcome"],drop_first=True,dtype="int64")
data
```

```python
x=data.drop("deposit",axis=1)
x
```

```python
y=data["deposit"]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2,shuffle=True)
print(f"x training {x_train.shape}")
print(f"y training {y_train.shape}")
print(f"x testing {x_test.shape}")
print(f"y testing {y_test.shape}")
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
bank_rf=RandomForestClassifier()
bank_rf
```

```python
bank_rf.fit(x_train,y_train)
y_predict=bank_rf.predict(x_test)
```

```python
y_predict
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_predict))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score
```

```python
select_model=[LogisticRegression(),
             DecisionTreeRegressor(max_depth=None,min_samples_leaf=2,min_samples_split=5),
             RandomForestRegressor(max_depth=None,min_samples_leaf=2,min_samples_split=5,n_estimators=100)]
```

```python
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings("ignore")
```

```python
for model in select_model:
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    print(f"accuracy {accuracy_score(y_pred,y_test)} of model is {model}")
```
